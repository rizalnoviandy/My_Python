```python
import pandas as pd
```

```python
cuaca = [    
    "mendung","kemarau","hujan",
    "hujan","hujan","kemarau",
    "hujan","kemarau","mendung",
        ]

suhu = [
    "dingin","panas","dingin",
    "dingin","dingin","panas",
    "dingin","panas","dingin",
]

windy = [
    True,False,True,
    True,False,True,
    True,False,False,
]

play = [
    "no","iya","iya",
    "iya","no","iya",
    "iya","","no",
]
```

```python
df = pd.DataFrame(
    {
        "kondisi": cuaca,
        "suhu": suhu,
        "windy": windy,
        "play": play,
    }
)
```
